Agrego el codigo de la regresion logistica para poder tener una base de donde salen las cosas

```python
import numpy as np
import pandas as np
```

```python

class RegresionLogistica:
    """ 
    Regresion logistica con descenso por gradiente #
    """
    
    def __init__(self) -> None:
        """
        Inicializacion del modelo
        : return: None
        """
        self.params = np.empty(0)
        self.cost_history = []
        self.params_history = []
        
    
    @staticmethod
    def sigmoide(z:float) ->float:
        """
        Funcion sigmoide
        :param z: input de la funcion sigmoide
        :return: resultado de la funcion sigmoide
        """
        
        if z >= 0:
            return 1/(1+np.exp(-z)) #Si z es muy grande y lo calculamos simplemente como exp(z) nos va a dar infinito y se va a romper todo
        else:                       #Con -z nos da 0 y nos va a dar bien
            return np.exp(z) / (1+np.exp(z)) #Y si z es muy chiquito (-z) se va a nan entonces usamos z solamente
        
        
        
    #LA FUNCION DE COSTO DE LA REGRESION PARA CLASIFICACION ES LA ENTROPIA CRUZADA   
    @staticmethod
    def cost_function(y:np.ndarray, y_pred:np.ndarray) -> float:
        """ 
        Funcion de costo de regresion logistica
        
        :param y: labels del conjunto
        :param y_pred: predicciones del modelo sobre el conjunto
        :return: costo del modelo
        """
        
        cost = 0
        
        for target,pred in zip(y,y_pred):  #zip es para iterar dos listas a la vez
            if target==1:
                cost += np.log(pred)
            else:
                cost += np.log(1-pred)
        return -1*cost
        
        
        
    #La tecnica que utilizabamos para calcular los patrones optimos era el descenso por gradiente, como el programa no para por si solo
    #Debemos darle algun parametro para que pare, eso lo podemos hacer con max_iter, que establece un maximo de iteraciones, tambien con
    #convergence_rate: la diferencia entre los parametros, si se alcanza se para.
    def fit(self, x_train:np.ndarray, y_train:np.ndarray, max_iter:int = 100000, convergence_rate: float = 0.00001, alpha = 0.0001) -> None:
        
        """ 
        Entrenamiento de la regresion logistica
        
        :param x_train: conjunto de entrenamiento
        :param y_train: labels de entrenamiento
        :param max_iter: cantidad maxima de iteraciones antes de detener el descenso por gradiente
        :param alpha: tasa de aprenzije (los pasos que da)
        """
        
        # np.hstack concatena matrices por columnas
        x = np.hstack((np.ones((x_train.shape[0], 1)), x_train))
        
        cost = []
        self.params = np.zeros(x.shape[1]) #Esto es para inicializar donde va a comenzar a calcular el gradiente, es random con 0s
        contador = 0 
        while contador < max_iter:
            contador += 1
            predicciones = np.array([self.sigmoide(np.dot(i, self.params.T)) for i in x])
        
            cost.append(self.cost_function(y_train, predicciones))
            vector_vertical = predicciones - y_train #La diferencia entre lo que predijimos y el valor real
            gradiente = np.dot(x.T, vector_vertical)       
            self.params_history.append(self.params.copy())
            old_params = self.params.copy()
            self.params = old_params - alpha*gradiente
        
            if np.linalg.norm(self.params - old_params) <= convergence_rate: #Si la diferencia entre los parametros y los parametros viejos es mas 
                break                                                         #chiquito que el convergence_Rate, cortamos el ciclo iterativo
        
        
        self.cost_history = cost 
        
        
        
    def get_params(self) -> np.ndarray:
        return self.params
    
 
    
    
```
